```python
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


# Definindo a lista de pontos a serem interpolados
X = [0.15,0.17,0.19,0.21,0.23,0.25,0.27,0.29,0.31]

# Calculando os valores de f(xi)
Y = [0.1761, 0.2304, 0.2788, 0.3222, 0.3617, 0.3979,0.4314, 0.4624, 0.4914]

# Criando uma lista de listas para a tabela de diferencas divididas 
dd=[Y] #ordem 0

# Gerando a tabela de dif. div. a partir da ordem 1 em diante
for o in range(1, len(X)):
    dd.append([])   # Adiciona uma lista vazia para armazenar as dds de ordem 1
    for k in range(0, len(X)-o, 1): 
        valor = (dd[o-1][k+1]-dd[o-1][k])/(X[k+o]-X[k])    
 
        dd[o].append(valor)
    


def produtorio(x,n):
    prod = 1.
    for i in range(n):
        prod = prod * (x-X[i])
    return prod


def calculaP(x):
    soma = dd[0][0]
    for i in range(1,len(X)):
        soma = soma + produtorio(x,i)*dd[i][0]
    return soma

print ("P(0.20) =", calculaP(0.20))
print ("P(0.22) =", calculaP(0.22))
 
# Cria a lista de pontos e calcula os valores para p plot
Xp = np.linspace(0.10, 0.40, num=10) 
Yp = []
for x in Xp:
    y = calculaP(x)
    Yp.append(y)

    
plt.plot(X, Y, 'o', Xp, Yp,'-', )
plt.grid()
plt.show()
```

```python
% matplotlib inline
import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt

X = np.array([50, 52, 54, 56, 58, 60]) # Distância da superfície
Y = np.array([2.75, 1.45,0.50, 0.15 ,0.20 ,0.85]) # Velocidade


# Calcula os elementos das marizes
A11 = np.sum(X**4)
A12 = np.sum(X**3)
A13 = np.sum(X**2)
A23 = np.sum(X)
A33 = len(X)
B1 = np.sum(X**2*Y) 
B2 = np.sum(X*Y)
B3 = np.sum(Y)

# Monta e resolve o sistema
A = np.array([[A11,A12, A13],[A12, A13, A23], [A13, A23, A33]])
B = np.array([B1,B2,B3])
a = solve(A, B)


# define a funcao g(x) para plotar 
g = lambda x: a[0]*x*x+a[1]*x+a[2]


# cria pontos (x, y) da reta 
Xr = np.arange(45, 65, 0.1)
Yr = []
for x in Xr:
    Yr.append(g(x))    

# Plota os pontos e a reta
plt.plot(X, Y, ".",Xr ,g(Xr), '-') 
#ponto de mínimo
xmin=-a[1]/(2*a[0])

print("θB=",x)
print("FB=",g(x))



plt.xlabel('Ângulo de incidência θ (°)',fontsize=16)
plt.ylabel('Fração da luz polarizada(%)',fontsize=16)
plt.plot(X,Y, "*", xmin, g(xmin),"ro") 
plt.show()
```
