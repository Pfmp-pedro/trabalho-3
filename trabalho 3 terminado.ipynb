```python
% matplotlib inline
import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt

X = np.array([50, 52, 54, 56, 58, 60]) # Distância da superfície
Y = np.array([2.75, 1.45,0.50, 0.15 ,0.20 ,0.85]) # Velocidade


# Calcula os elementos das marizes
A11 = np.sum(X**4)
A12 = np.sum(X**3)
A13 = np.sum(X**2)
A23 = np.sum(X)
A33 = len(X)
B1 = np.sum(X**2*Y) 
B2 = np.sum(X*Y)
B3 = np.sum(Y)

# Monta e resolve o sistema
A = np.array([[A11,A12, A13],[A12, A13, A23], [A13, A23, A33]])
B = np.array([B1,B2,B3])
a = solve(A, B)


# define a funcao g(x) para plotar 
g = lambda x: a[0]*x*x+a[1]*x+a[2]


# cria pontos (x, y) da reta 
Xr = np.arange(45, 65, 0.1)
Yr = []
for x in Xr:
    Yr.append(g(x))    

# Plota os pontos e a reta
plt.plot(X, Y, ".",Xr ,g(Xr), '-') 
#ponto de mínimo
xmin=-a[1]/(2*a[0])

print("Ângulo de incidência θ (°)=",x)
print("Fração da luz polarizada(%)=",g(x))



plt.xlabel('Ângulo de incidência θ (°)',fontsize=16)
plt.ylabel('Fração da luz polarizada(%)',fontsize=16)
plt.plot(X,Y, "*", xmin, g(xmin),"ro") 
plt.show()
```

### Questão 4 b:
Para prever µ de 7,5°C usamos o método de interpolação de Lagrange.

```python
% matplotlib inline
import matplotlib.pyplot as plt

X = np.array([0, 5, 10, 20, 30, 40])
Y = np.array([1.787, 1.519, 1.307, 1.002, 0.7975, 0.6529])

n = len(Y)

def calculaP(x):
    valor = 0
    for k in range(n):
        lk = 1.0
        for i in range(n):
            if k != i:
                lk = lk*(x - X[i])/(X[k]-X[i])
        valor = valor + Y[k]*lk 
    return valor

Ynew = []
Xnew = np.linspace(X[0], X[-1], num=21)
for xnew in Xnew:
    ynew = calculaP(xnew)
    Ynew.append(ynew)

plt.plot(X, Y, 'o', Xnew, Ynew,'-',x,calculaP(x),'ro')
plt.grid()
plt.show()

x = 7.5
print ("µ(",x,") =", calculaP(x))
```

A partir dos valores dados de Tensão e taxa de arrasto, foi possível linearizar os pontos através do método de ajuste de potência, onde aplicamos ln tanto nos dados de arrasto quanto nos dados da tensão, logo depois, criamos a funçao f(x) e usamos a função Solve do scipy. Após isso, foi possível encontrar os valores de β e de m.
Com os valores de β e m em mãos, criamos uma função g(x) onde os x são os valores de tensão e, através desta função, foi possível uma boa aproximação dos valores da taxa de arrasto.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import solve
import math

X = np.array([5.775, 8.577, 10.874, 12.555])
Y = np.array([0.0004, 0.0011, 0.0021, 0.0031])
X1 = np.log(X)
Y1 = np.log(Y)

# Calcula os elementos das matrizes
a11 = np.sum(X1**2)
a12 = np.sum(X1)
a22 = len(X1)
b1 = np.sum(X1 * Y1)
b2 = np.sum(Y1)

# Monta e resolve o sistema
A = np.array([[a11, a12], [a12, a22]])
B = np.array([b1,b2])
a = solve(A, B)

# define a funcao g(x) para plotar 
f = lambda x: a[0]*x+a[1]

# cria pontos (x, y) da reta 
Xr = np.linspace(X1[0],X1[-1],num=50)

# Função Linearizada
plt.plot(X1, Y1, ".", Xr, f(Xr), "-") 
plt.show()

#Encontrando valores para β e m
β = math.exp(a[1])
m = a[0]
print("O Valor de β é: ",β)
print("O Valor de m é: ",m)

g = lambda x: β*x**m

Xn = np.linspace(X[0],X[-1],num=100)
Yn = []
for x in Xn:
    Yn.append(g(x))


plt.plot(X, Y, ".", Xn, Yn, "r-")
plt.show()
```
